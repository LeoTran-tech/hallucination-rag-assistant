```python
from pathlib import Path
import random
import fitz # is the PyMuPDF library, used for working with PDF files

PDF_DIR = Path("../data/raw/papers")

# Find all PDF files in the specified directory
pdf_files = list(PDF_DIR.glob("*.pdf"))

print("Directory exists:", PDF_DIR.exists())
print("Total PDFs found:", len(pdf_files))

# Randomly sample a few PDF files
sample_files = random.sample(pdf_files, min(5, len(pdf_files)))
sample_files
```

```python
# Function to extract text preview from a PDF
def extract_text_preview(pdf_path):

    # Open the PDF document
    doc = fitz.open(pdf_path)

    # Define a list to hold text previews
    texts = []

    # Loop through all pages
    for page_num, page in enumerate(doc):

        # Extract text from the page
        text = page.get_text("text")

        # Store the page number and the preview of the page text
        texts.append((page_num + 1, text[:500]))  # preview 500 chars
    doc.close()
    return texts

# Loop throught the sampled PDF files
for pdf in sample_files:
    print("="*80)
    print("FILE:", pdf.name)
    previews = extract_text_preview(pdf)

    # Print the preview of the first page of the PDF
    print("Page 1 preview:\n", previews[0][1])
```
